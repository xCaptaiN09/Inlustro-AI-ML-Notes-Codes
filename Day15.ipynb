```python
#Bernouli Distribution - p^x(1 -p)^1-x
#probability Distribution
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
p = 0.7 #probability of success 
x = [0,1] #possible outcomes
# probability mass function (pmf)
#probability density function (pdf)
pmf = bernoulli.pmf(x, p)
plt.scatter(x,pmf,color="black")
plt.title("bernouli distribution")
plt.xlabel("bernoyli's pmf value")
plt.ylabel("the possible outcomes")
plt.show()

```

```python
#Binomial distribution - nc^xp^xq^(n-x)
'''
n,p,q,x
'''
from scipy.stats import binom
import numpy as np
n = 10
p = 0.7 #q=1-P
x = np.arange(0,n+1)
pmf = binom.pmf(x,n,p)
plt.scatter(x,pmf,color="red")
plt.title("Binomial distribution")
plt.xlabel("the pmf value is")
plt.ylabel("the outcome is")
plt.show()
```

```python
#Posission Distribution
'''
to calculate the average possible outcomes or rates
with in the interval of time e^-lambda.lambda^k/n!

'''

from scipy.stats import poisson
lambda_value = 5
k = np.arange(0,10)
pmf = poisson.pmf(lambda_value,k)
print(pmf)
plt.bar(k,pmf,color="red")
plt.title("call range")
plt.xlabel("daily range")
plt.ylabel("usage")
plt.show()
```

```python
#continuos - 
#1.uniform Distribution - 1/b-a
from scipy.stats import uniform
a = 2
b = 5
x =np.linspace(a,b,200)
pdf = uniform.pdf(x,loc=a,scale=b-a) #loc is  
print(pdf)
plt.bar(x,pdf,color="maroon")
plt.show()
```

```python
#Exponential - average rate(lambda.e^-lambda)
from scipy.stats import expon
lambda_value = 2
x=np.linspace(0,10,50)
pdf = expon.pdf(x,scale=1/lambda_value)
print(pdf)
plt.bar(x,pdf,color="violet")
plt.show()
```

```python
def mean_variance(x,prob_x):
    E = sum(V*p for V,p in zip(x,prob_x))
    E2 = sum((V**2)*p for V,p in zip(x,prob_x))
    #for V,p in zip(x,prob_x):
        #s1 = sum(V)*p
    #for V,p in zip(x,prob_x):
        #s2 = sum(V**2)*p
    var_value = E2 - E**2
    return E , var_value
x = [0,1,2]
prob_x = [0.3,0.5,0.2]
E , V = mean_variance(x,prob_x)
print(f"The mean value is {E} and the Variance is {V}")


```

```python
#Central limit theorem
import numpy as np
import matplotlib.pyplot as plt

n_trials = 1000
n_dice = 30
sample_means = []
for i in range(n_trials):
    rolls = np.random.randint(1,7,n_dice)
    sample_means.append(np.mean(rolls))

plt.hist(sample_means, bins=30, color= 'skyblue',edgecolor='black',density=True)
plt.title('Central Limit Theorem with Dice Rolls')
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.axvline(x=3.5,color='red',linestyle='',label='Expected mean(3.5)')
plt.legend()
plt.show()
```
